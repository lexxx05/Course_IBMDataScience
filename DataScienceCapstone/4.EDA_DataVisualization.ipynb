```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Using the ready dataset from the previous activities
df=pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_2.csv")
df.head(5)
```

```python
# Plot the flight number vs payload mass
sns.catplot(y='PayloadMass', x='FlightNumber', hue = 'Class', data = df, aspect = 5)
plt.xlabel("Flight Number",fontsize=20)
plt.ylabel("Pay load Mass (kg)",fontsize=20)
plt.show()
```

1. Relationship between Flight Number and Launch Site

```python
sns.catplot(x='FlightNumber', y='LaunchSite', hue='Class', data = df)
plt.xlabel("Flight Number",fontsize=20)
plt.ylabel("Launch Site",fontsize=20)
plt.show()
```

2. Relationship between Payload and Launch Site`

```python
sns.catplot(x='PayloadMass', y='LaunchSite', hue='Class', data = df)
plt.xlabel("Payload Mass",fontsize=20)
plt.ylabel("Launch Site",fontsize=20)
plt.show()
```

3. Relationship between Success Rate of Each Orbit Type

```python
dfgrouped = df.groupby(['Orbit'])['Class'].mean().reset_index()
sns.barplot(x='Orbit', y='Class', data = dfgrouped)
plt.xlabel("Orbit",fontsize=20)
plt.ylabel("Class",fontsize=20)
plt.show()
```

4. Relationship between Flight Number and Orbit Type

```python
sns.catplot(x='FlightNumber', y='Orbit', hue='Class', data = df)
plt.xlabel("Flight Number",fontsize=20)
plt.ylabel("Orbit",fontsize=20)
plt.show()
```

5. Relationship between Pay load and Orbit Type

```python
sns.catplot(x='PayloadMass', y='Orbit', hue='Class', data = df)
plt.xlabel("PayloadMass",fontsize=20)
plt.ylabel("Orbit",fontsize=20)
plt.show()
```

6. Yearly Trend of Launch Successes

```python
# Function to extract year from the date
year = []
def extractYear(data):
  for i in df['Date']:
    year.append(i.split('-')[0])
  return year

extractYear(df)
df['year'] = year
```

```python
dfgrouped = df.groupby(['year'])['Class'].mean().reset_index()
sns.lineplot(x='year', y='Class', data=df)
plt.xlabel("Year")
plt.ylabel("Average Success Rate")
plt.show()
```
