```python
!pip install dash
import numpy as np
import pandas as pd
import plotly.express as px
import dash
from dash import dcc
from dash import html
from dash.dependencies import Input, Output
```

```python
dfs =  pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/airline_data.csv',
                       encoding = "ISO-8859-1",
                       dtype={'Div1Airport': str, 'Div1TailNum': str,
                              'Div2Airport': str, 'Div2TailNum': str})

app = dash.Dash(__name__)

app.layout = html.Div(children = [
    html.H1(
        "Total Number of Fligts to the Destination State Split by Reporting Air",
            style = {
                'textAlign':'center',
                'color' : '#FFFFFF',
                'font-size' : 50
            }
        ),

    html.Div(
        [
            'Input Your Selected Year :  ',
            dcc.Input(
                id = 'inputYear',
                value = '2010',
                type = 'number',
                style = {
                    'height' : '50px',
                    'font-size' : 35
                }
            ),
        ], style = {'font-size':40, 'color' : '#FFFFFF'}
    ),

    html.Br(),
    html.Br(),
    html.Div(dcc.Graph(id = 'barChart')),
]
                      )


@app.callback(
    Output(component_id = 'barChart', component_property = 'figure'),
    Input(component_id = 'inputYear', component_property = 'value'),
)

def getGraph(enteredYear):
  dfss = dfs[dfs['Year'] == int(enteredYear)]
  barDf = dfss.groupby('DestState')['Flights'].sum().reset_index()
  fig = px.bar(barDf, x = 'DestState', y = 'Flights', title='Total number of flights to the destination state split by reporting airline')
  fig.update_layout(title = 'Flights to Destination State', xaxis_title = 'DestState', yaxis_title = 'Flights')
  return fig

if __name__ == '__main__':
  app.run_server()
```
