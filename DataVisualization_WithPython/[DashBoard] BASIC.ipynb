```python
#import the required libraries
import numpy as np
import pandas as pd
import plotly.express as px
import dash
from dash import dcc
from dash import html
```

# 1. Simple Dash

## 1.1. Prepare and Visualize Data

```python
#Prepare the data
df = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/airline_data.csv',
                            encoding = "ISO-8859-1",
                            dtype={'Div1Airport': str, 'Div1TailNum': str,
                                   'Div2Airport': str, 'Div2TailNum': str})

#randomly sample 500 data
df = df.sample(n=500, random_state = 42)

#create the interactive visualization of pie charts
ax = px.pie(df, values = 'Flights', names = 'DistanceGroup', title = 'Distance Group Proportion by Flights')
ax.show()
```

## 1.2. Create the Base of Dash Application

```python
#create a dash application
apps = dash.Dash(__name__)

#get the layout of application n adjust it
#creat outer division using html.div and add title to the dashboard
#add description about the graph using HTML P componen (paragraph)
#finally, add graph component

apps.layout = html.Div(
    children = [html.H1(), #title
                html.P(), #description
                dcc.Graph()] #chart
)

#run the application
if __name__ == '__main__':
  apps.run_server()
```

## 1.3. Update the Title, Description and Chart

```python
apps = dash.Dash(__name__)

apps.layout = html.Div(
    children = [html.H1('Airline Dashboard',
                        style = {'textAlign' : 'center',
                                'color' : '#FFFFFF',
                                'font-size' : 40}), #1. update the title
                html.P('Proportion of distance group (250 mile distance interval group) by flights.',
                       style={'textAlign':'center',
                              'color': '#F57241'}), #2. update the description
                dcc.Graph(figure = ax)] #chart
)

#run the application
if __name__ == '__main__':
  apps.run_server()
```
